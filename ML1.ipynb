```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('wine.csv')
```

```python
df
```

```python
# Create the scatter plot
x_feature1 = 'Alcohol'
x_feature2 = 'Malic_Acid'
y = df['Customer_Segment']
plt.figure(figsize=(8, 6))
plt.scatter(df[x_feature1], df[x_feature2], c=y, cmap='viridis', edgecolor='k')
plt.colorbar(label='Customer Segment')
plt.xlabel(x_feature1)
plt.ylabel(x_feature2)
plt.title("Scatter Plot of Alcohol vs Malic_Acid Before PCA")
plt.show()
```

```python
x = df.drop(columns=['Customer_Segment'])
y = df['Customer_Segment']
pipe=Pipeline([('scaler',StandardScaler()),('pca',PCA(n_components=2))])
```

```python
xt=pipe.fit_transform(x)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(xt[:,0],xt[:,1],c=y)
plt.legend(labels=list(df['Customer_Segment']))
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("First two principal components after scaling")
plt.show()
```
